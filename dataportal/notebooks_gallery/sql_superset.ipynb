Kjør SQL via SuperSet SQL Lab eller direkte fra Python. Dette er utgangspunktet for et dashboard.

## Alternativ A: SQL Lab i Superset

Åpne [Superset SQL Lab](http://localhost:8088/sqllab) og kjør:

```python
# Markdown-celle viser SQL — ikke kjørbar her, men kan limes inn i SQL Lab:
SQL = """
SELECT domain, COUNT(*) AS rows
FROM information_schema.tables
WHERE table_schema = 'gold'
GROUP BY domain
ORDER BY rows DESC;
"""
print(SQL)
```
